```python
np.random.seed(0)
# For each FPL Manager

fpl_manager_id = '2757'

# Get the players picked by a FPL Manager in week 1
# player_ids = []
r = requests.get('https://fantasy.premierleague.com/api/entry/'+fpl_manager_id+'/event/1/picks/').json()
player_ids = [x['element'] for x in r['picks']]

player_ids

scores_each_week = []
team_each_week = []

#Initialise total team points
df_tmp = pd.read_csv("Players_Weekwise/week_1.csv")
df_tmp = df_tmp[['id', 'stats.total_points']]
df_tmp = df_tmp.set_index('id')
total_points = df_tmp.loc[player_ids, :]['stats.total_points'].sum()

scores_each_week.append(total_points)
team_each_week.append(player_ids)

for week in range(2,8):
    # Get number of players in that week available for swap
    df = pd.read_csv("Players_Weekwise/week_"+str(week)+".csv")
    df = df[['id', 'stats.total_points']]
    df = df.set_index('id')
    # print(df)
    
#     total_points = df.loc[player_ids, :]['stats.total_points'].sum()
        
    print("Total Points "+str(total_points))
    # Replicate player scores horizontally len(df) times
    scores_h = np.array([df.loc[player_ids, :]['stats.total_points'].to_numpy()])
    scores_h = np.repeat(scores_h, len(df), axis = 0)
    scores_h = scores_h.T
#     print(scores_h)
    
    # Get all player_scores into matrix form.
    ppoints = np.array([df['stats.total_points'].to_numpy()])
    ppoints = np.repeat(ppoints, 15, axis=0)
    
    #Create total score matrix
    total_score = np.full((15,len(df)), total_points)
    
    #Create transfers/swaps to get new scores
    new_team_scores = total_score + ppoints - scores_h
#     print(new_team_scores)
    
    # Randomly CROSS OUT some transfers based on some cost
    random_cost = np.random.choice([0,1],(15, len(df)), p=[0.5, 0.5])
    new_score_with_cost = np.multiply(random_cost, new_team_scores)
#     print(new_score_with_cost)
    # Row Number. Player swapped OUT.
#     print(np.argmax(np.max(new_score_with_cost, axis=1)))
    # Row Number. Player swapped IN.
#     print(np.argmax(np.max(new_score_with_cost, axis=0)))
    
#     print(np.max(new_score_with_cost))
    
    print("Swap OUT")
    player_id_swap_out = player_ids[np.argmax(np.max(new_score_with_cost, axis=1))]
    print(df.loc[player_id_swap_out]['stats.total_points'])
#     print(player_id_swap_out)
    
#     print(scores_h[np.argmax(np.max(new_score_with_cost, axis=1))])
    
    print("Swap IN")
    player_id_swap_in = df.reset_index()['id'][np.argmax(np.max(new_score_with_cost, axis=0))]
#     print(player_id_swap_in)
    print(df.loc[player_id_swap_in]['stats.total_points'])
#     print(ppoints[0][np.argmax(np.max(new_score_with_cost, axis=0))])
    
    player_ids.extend([player_id_swap_in])
    player_ids.remove(player_id_swap_out)
    print("New Points "+ str(np.max(new_score_with_cost)))
    scores_each_week.append(np.max(new_score_with_cost))
    new_player_ids = player_ids
    team_each_week.append(new_player_ids)
    total_points = np.max(new_score_with_cost)
    print(new_player_ids)
```
