```python
results = []
for profile in profiles_2:
    results.append(list(df_player_new.sort_values(by=profile['cols'], ascending=profile['order']).index))
```

```python
def k_intersection(l1, l2, k):
    return set(l1[:k]).intersection(l2[:k])
```

```python
def common(results):
    top_k = [50,100,125,150,200]
    for k in top_k:
        for i,l1 in enumerate(results):
            for j,l2 in enumerate(results):
                if i!= j :
                    print("Top "+str(k))
                    print(profiles_2[i]['cols'])
                    print(profiles_2[j]['cols'])
                    print("Common " + str(len(k_intersection(l1, l2, k))))
                    print("----------------------------")
```
